```python
# plotting a bar chart for Category and it's count
ax = sns .countplot(x = 'Category' , data =df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# plotting a bar chart for Sub-Category and it's count
ax = sns .countplot(x = 'Sub-Category' , data =df)

for bars in ax.containers:
    ax.bar_label(bars)
    sns.set(rc={'figure.figsize':(25,5)})
```

```python
df.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.title('Total Profit and Sales per Sub-Category')
# plt.legend('Profit')
# plt.legend('Sales')
plt.show()
```

```python
#3.1 Sales analysis
```

```python
# Total sales per category
sales_per_category = df.groupby('Category')['Sales'].sum().reset_index()
```

```python
# plotbar chart for total sales per category
plt.figure(figsize=(10, 6))
plt.bar(sales_per_category['Category'], sales_per_category['Sales'])
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.title('Total Sales per Category')
plt.xticks(rotation=45)
plt.show()
```

```python
# Total sales per region
sales_per_region = df.groupby('Region')['Sales'].sum().reset_index()
```

```python
# Pie chart for total sales per region
plt.figure(figsize=(8, 8))
plt.pie(sales_per_region['Sales'], labels=sales_per_region['Region'], autopct='%1.1f%%')
plt.title('Total Sales per Region')
plt.show()
```

```python
# Top 10 selling products
top_10_products = df.groupby('Product Name')['Quantity'].sum().nlargest(10).reset_index()
```

```python
#  plotbar chart  for top 10 selling products
plt.figure(figsize=(12, 6))
plt.bar(top_10_products['Product Name'], top_10_products['Quantity'])
plt.xlabel('Product Name')
plt.ylabel('Quantity Sold')
plt.title('Top 10 Selling Products')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Calculate total sales by each customer
total_sales_by_customer = df.groupby('Customer Name')['Sales'].sum().reset_index()
```

```python
# Visualize top 10 customers with the highest sales
plt.figure(figsize=(10, 6))
plt.bar(total_sales_by_customer['Customer Name'].head(10), total_sales_by_customer['Sales'].head(10))
plt.xlabel('Customer Name')
plt.ylabel('Total Sales')
plt.title('Top 10 Customers by Total Sales')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Calculate the number of orders made by each customer
order_count_by_customer = df.groupby('Customer Name')['Order ID'].nunique().reset_index()
order_count_by_customer = order_count_by_customer.sort_values(by='Order ID', ascending=False)
```

```python
# Visualize top 10 customers with the highest number of orders
plt.figure(figsize=(10, 6))
plt.bar(order_count_by_customer['Customer Name'].head(10), order_count_by_customer['Order ID'].head(10))
plt.xlabel('Customer Name')
plt.ylabel('Number of Orders')
plt.title('Top 10 Customers by Number of Orders')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Total profit per category
profit_per_category = df.groupby('Category')['Profit'].sum().reset_index()
```

```python
#  plotbar for total profit per category
plt.figure(figsize=(10, 6))
plt.bar(profit_per_category['Category'], profit_per_category['Profit'])
plt.xlabel('Category')
plt.ylabel('Total Profit')
plt.title('Total Profit per Category')
plt.xticks(rotation=45)
plt.show()
```

```python
#Profit Analysis
```

```python
# Total profit per region
profit_per_region = df.groupby('Region')['Profit'].sum().reset_index()
```

```python
# Pie chart for total profit per region
plt.figure(figsize=(8, 8))
plt.pie(profit_per_region['Profit'], labels=profit_per_region['Region'], autopct='%1.1f%%')
plt.title('Total Profit per Region')
plt.show()
```

```python
# Top 10 most profitable products
top_10_products = df.groupby('Product Name')['Profit'].sum().nlargest(10).reset_index()
```

```python
# Bar plot for top 10 most profitable products
plt.figure(figsize=(12, 6))
plt.bar(top_10_products['Product Name'], top_10_products['Profit'])
plt.xlabel('Product Name')
plt.ylabel('Profit')
plt.title('Top 10 Most Profitable Products')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
